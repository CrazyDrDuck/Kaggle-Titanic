```python
df_titanic["Relatives"] = (df_titanic["SibSp"] + df_titanic["Parch"])
```

```python
df_titanic["Pclass"]=df_titanic["Pclass"].astype("int")
```

```python
for klass in set(df_titanic.Pclass):
    df_titanic.loc[df_titanic.Pclass==klass,"Age_filled"]=df_titanic.loc[df_titanic.Pclass==klass,"Age"].fillna(df_titanic.loc[df_titanic.Pclass==klass].Age.mean())
```

```python
df_titanic["Sex"] = df_titanic["Sex"].astype("category")
```

```python
df_titanic["Sex"]=df_titanic["Sex"].cat.codes
```

```python
df_titanic["Sex"]
```

```python
df_titanic["Fare_adjusted"] = df_titanic["Fare"] / (df_titanic["Relatives"]+1)
```

```python
df_titanic["Deck"]=df_titanic.Cabin.str[0]
```

```python
df_titanic.loc[df_titanic.Pclass==1, "Deck"] = df_titanic.fillna("B")
```

```python
df_titanic.loc[df_titanic.Pclass==2, "Deck"] = df_titanic.fillna("D")
```

```python
df_titanic.loc[df_titanic.Pclass==3, "Deck"] = df_titanic.fillna("F")
```

```python
df_titanic["Deck"]=df_titanic["Deck"].astype("category")
```

```python
df_titanic["Deck"]=df_titanic["Deck"].cat.codes
```

```python
y = df_titanic["Survived"]
```

```python
y
```

```python
X = df_titanic[["Age_filled", "Sex", "Relatives", "Pclass", "Fare_adjusted", "Deck"]].copy()
```

```python
X
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
dftp = LogisticRegression(max_iter=500)
```

```python
dftp.fit(X=X_train, y=y_train)
```

```python
X_test
```

```python
y_test
```

```python
dftp.score(X=X_test, y=y_test, sample_weight=None)
```

```python
dftp.coef_
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(dftp, X, y, cv=20)
```

```python
scores
```

```python
scores.mean()
```

```python
y_pred = dftp.predict(X_test)
```

```python
pd.crosstab(y_test, y_pred)
```
